## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana rata-rata kandungan Karbon Monoksida (CO) di udara dari tahun 2013-2017?

```python
# Group by year dan station, hitung rata-rata CO setiap year dan setiap station
mean_co_by_year_station = combined_df.groupby(['year', 'station'])['CO'].mean().unstack()


plt.figure(figsize=(10, 5))

# Plot 3 garis untuk 3 station
for station in mean_co_by_year_station.columns:
    plt.plot(mean_co_by_year_station.index, mean_co_by_year_station[station], marker='o', linewidth=2, label=station)
    
    for i, txt in enumerate(mean_co_by_year_station[station]):
        plt.text(mean_co_by_year_station.index[i], txt, f'{txt:.2f}', ha='center', va='bottom', fontsize=8)

plt.title("Rataan Kandungan CO di Udara tahun 2013-2017", loc="center", fontsize=20)
plt.xlabel("Year", fontsize=12)
plt.ylabel("RerataCO", fontsize=12)
plt.legend()  
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 2: Bagaimana tren rata-rata temperatur bulanan untuk all year 2013-2017?

```python
# Group by month dan station, hitung rata-rata CO setiap month dan setiap station
mean_temp_by_month_station = combined_df.groupby(['month', 'station'])['TEMP'].mean().unstack()


plt.figure(figsize=(10, 5))

# Plot 3 garis untuk 3 station
for station in mean_temp_by_month_station.columns:
    plt.plot(mean_temp_by_month_station.index, mean_temp_by_month_station[station], marker='o', linewidth=2, label=station)
    
    for i, txt in enumerate(mean_temp_by_month_station[station]):
        plt.text(mean_temp_by_month_station.index[i], txt, f'{txt:.2f}', ha='center', va='top', fontsize=8)

plt.title("Rataan TEMP bulanan untuk all year 2013-2017", loc="center", fontsize=20)
plt.xlabel("Month", fontsize=12)
plt.ylabel("RerataTEMP", fontsize=12)
plt.legend()  
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 3: Bagaimana tren rata-rata curah hujan bulanan untuk all year 2013-2017?

```python
# Group by month dan station, hitung rata-rata curah hujan setiap month dan setiap station
mean_rain_by_month_station = combined_df.groupby(['month', 'station'])['RAIN'].mean().unstack()


plt.figure(figsize=(10, 5))

# Plot 3 garis untuk 3 station
for station in mean_rain_by_month_station.columns:
    plt.plot(mean_rain_by_month_station.index, mean_rain_by_month_station[station], marker='o', linewidth=2, label=station)
    
    for i, txt in enumerate(mean_rain_by_month_station[station]):
        plt.text(mean_rain_by_month_station.index[i], txt, f'{txt:.2f}', ha='center', va='top', fontsize=8)

plt.title("Rataan curah hujan bulanan untuk all year 2013-2017", loc="center", fontsize=20)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Rerata curah hujan", fontsize=12)
plt.legend()  
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Conclusion

- Conclution pertanyaan 1: Rata-rata kandungan CO di udara fluktuatif sepanjang 2013-2017 di ketiga daerah, kandungan CO mencapai puncaknya di tahun 2017 terutama meningkat tajam di Gucheng.
- Conclution pertanyaan 2: Rata-rata temperatur di ketiga daerah sepanjang 2013-2017 meningkat dengan puncak di bulan Juli setiap tahunnya atau bertepatan dengan musim kemarau. Rata-rata temperatur akan mencapai titik terdingin di bulan Januari setiap tahunnya atau bertepatan dengan musim dingin.
- Conclution pertanyaan 3: Rata-rata curah hujan mencapai puncaknya di bulan Juli setiap tahunnya. Rata-rata curah hujan mencapai titik terendah di Januari dan Desember karena tidak terdapat hujan air.
